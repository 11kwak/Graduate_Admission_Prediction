```python
import tensorflow as tf
import pandas as pd
import numpy as np

df = pd.read_csv("./data/gpascore.csv", encoding="cp949")

# df.isnull().sum() # 결측치 확인  gre      1
# df.fillna()  # 빈칸 채워줌 
df = df.dropna()  #결측치 행 삭제
# print(df['gpa'].min()) 최저확인
print(df)





```

```python
y데이터 = df["admit"].values

x데이터 = []
for i,rows in df.iterrows():      #pandas iterrow 함수 한행씩 출력하겟금
    x데이터.append([ rows['gre'], rows['gpa'], rows['rank']   ])
```

```python
model = tf.keras.models.Sequential([
tf.keras.layers.Dense(64, activation="tanh"), 
tf.keras.layers.Dense(128, activation="tanh"),
tf.keras.layers.Dense(1, activation="sigmoid"), # 결과가 0과 1사이 분류 면 sigmoid 0 or 1 뱉어야 하니 , 마지막 레이어 에서 덴스가 하나 
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) #binary_crossentropy 확률문제에서 loss함수는 이거씀

model.fit(np.array(x데이터), np.array(y데이터), epochs=1000) #numpy 나 tf 형식으로 학습 
```

```python
예측값 = model.predict( [[860,3.76,3], [920,3.76,1], [990,3.76,1]] )
print(예측값)
```
